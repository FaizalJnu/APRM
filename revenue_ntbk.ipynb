```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from datetime import datetime
from imblearn.under_sampling import RandomUnderSampler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
Data_12month_revenue = pd.read_csv('12months_flight_revenue.csv')
Data_3month_revenue = pd.read_csv('3months_flight_revenue_prediction.csv')
```

```python
pd.set_option('display.max_columns', None)
print(Data_3month_revenue)
```

```python
# For 'Cabin' column in Data_12month_revenue
cabin_mapping = {'Y': 1, 'J': 2}  # Map 'Y' to 1
Data_12month_revenue['Cabin'] = Data_12month_revenue['Cabin'].map(cabin_mapping)

label_encoder = LabelEncoder()
Data_12month_revenue['Booking Class'] = label_encoder.fit_transform(Data_12month_revenue['Booking Class'])

# For 'Origin' and 'Destination' columns in Data_12month_revenue
Data_12month_revenue[' Origin'] = label_encoder.fit_transform(Data_12month_revenue[' Origin'])
Data_12month_revenue[' Destination'] = label_encoder.fit_transform(Data_12month_revenue[' Destination'])

# Convert 'Dep_Date' to a numerical format (e.g., number of days since a specific date)
# Assuming the date is in the format 'YYYY-MM-DD'
Data_12month_revenue['Dep_Date'] = pd.to_datetime(Data_12month_revenue['Dep_Date'])
Data_12month_revenue['Dep_Date_Numerical'] = (Data_12month_revenue['Dep_Date'] - pd.to_datetime('2022-01-01')).dt.days

# Split 'Dep_Time' into hours and minutes
Data_12month_revenue['Dep_Hour'] = Data_12month_revenue['Dep_time'].str.split(':').str[0].astype(int)
Data_12month_revenue['Dep_Minute'] = Data_12month_revenue['Dep_time'].str.split(':').str[1].astype(int)


# Drop the original 'Dep_Date' and 'Dep_Time' columns
Data_12month_revenue = Data_12month_revenue.drop(['Dep_Date', 'Dep_time'], axis=1)

# Drop columns with all NaN values
Data_12month_revenue.dropna(axis=1, how='all', inplace=True)
# Calculate the threshold for dropping columns
threshold = len(Data_12month_revenue) * 0.05  # 5% threshold
Data_12month_revenue.dropna(axis=1, thresh=threshold, inplace=True)
Data_12month_revenue.fillna(0.00, inplace=True)
```
